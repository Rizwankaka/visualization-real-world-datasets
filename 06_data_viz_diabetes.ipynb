# Data Visulization on 'Diabetes' dataset using `Pandas`

- The dataset can be downloaded from Kaggle [link](https://www.kaggle.com/datasets/mathchi/diabetes-data-set)

## About the Dataset

**Context**

This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective is to predict based on diagnostic measurements whether a patient has diabetes.

**Content**

Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

- Pregnancies: Number of times pregnant
- Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- BloodPressure: Diastolic blood pressure (mm Hg)
- SkinThickness: Triceps skin fold thickness (mm)
-  nsulin: 2-Hour serum insulin (mu U/ml)
- BMI: Body mass index (weight in kg/(height in m)^2)
- DiabetesPedigreeFunction: Diabetes pedigree function
- Age: Age (years)
- Outcome: Class variable (0 or 1)

```python
# import the libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the diabetes dataset
df = pd.read_csv('./data/diabetes.csv')
```

### Plot1. Histogram

```python
# 1. Histogram of glucose levels

# Set the figure size
plt.figure(figsize=(10,6))

# Plot the histogram of glucose levels
plt.hist(df['Glucose'], bins=20, color='green')

# Set the x and y-axis labels 
plt.xlabel('Glucose Level'); plt.ylabel('Frequency')

# Set the plot title
plt.title('Distribution of Glucose Levels')

# Save the figure as a png file with 300 dpi resolution
plt.savefig('./outputs/diabetes_hist.png', dpi = 300)

# Display the plot
plt.show()
```

**Interpretation:**
- Glucose levels are right skewed with most values less than 200

### Plot 2. Scatterplot

```python
# 2. Scatterplot of glucose vs BMI

# Set the figure size
plt.figure(figsize=(10,6))

# Plot the scatterplot
plt.scatter(df['BMI'], df['Glucose'], color='red')

# Set the x and y-axis labels
plt.xlabel('BMI'); plt.ylabel('Glucose')

# Set the plot title
plt.title('Glucose vs BMI')

# Save the figure as a png file with 300 dpi resolution
plt.savefig('./outputs/diabetes_scatter.png', dpi = 300)

# Display the plot
plt.show()
```

**Interpretation:**

- Positive correlation between BMI and glucose levels

### Plot 3. Count Plot

```python
# 3. Countplot of outcome

# Set the figure size
plt.figure(figsize=(10,6))

# Plot the countplot of outcome
sns.countplot(df['Outcome'], color='blue')

# Set the x-axis label
plt.xticks([0,1], ['Negative', 'Positive'])

# Set the plot title
plt.title('Outcome Frequency')

# Save the plot 
plt.savefig('./outputs/diabetes_countplot.png', dpi = 300)

# Show the plot 
plt.show()
```

**Interpretation**
- Imbalanced classes - majority negative outcomes

### Plot4. Boxplot

```python
# 4. Boxplot of glucose by outcome

# Set the figure size
plt.figure(figsize=(10,6))

# Plot the boxplot of glucose by outcome
df.boxplot(column='Glucose', by='Outcome')

# Set the plot title
plt.title('Glucose Distribution by Outcome')

# Save the plot
plt.savefig('./outputs/diabetes_boxplot.png', dpi = 300)

# Show the plot
```

**Interpretation**
- Those with positive outcome tend to have higher glucose

### Plot 5. Violin Plot 

```python
# 5. Violin plot of BMI by outcome

# Set the figure size
plt.figure(figsize=(10,6))

# plot the violin plot of BMI by outcome
sns.violinplot(x='Outcome', y='BMI', data=df)

# Set the plot title
plt.title('BMI Distribution by Outcome')

# Save the plot
plt.savefig('./outputs/diabetes_violinplot.png', dpi = 300)

# Show the plot
plt.show()
```

**Interpretation**
BMI distribution very similar for both outcomes

### Plot 6. Jointplot

```python
# 6. Jointplot of BMI vs glucose

# Set the figure size
plt.figure(figsize=(10,6))

# Plot the Jointplot 
sns.jointplot(data=df, x='BMI', y='Glucose', kind='reg', color='purple')

# Set the plot title
plt.title('BMI vs Glucose')

# Save the plot 
plt.savefig('./outputs/diabetes_jointnplot.png', dpi = 300)

# Display the plot
plt.show()
```

**Interpretation**
- Linear relationship between BMI and glucose


### Plot 7. Pairplot of features 

```python
# Pairplot of features

import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10,6)) 

# Plot the pairplot
sns.pairplot(df[['Pregnancies','Glucose','BloodPressure','SkinThickness','BMI']])

# Set the plot title  
plt.title('Pairplot of Features')

# Save the plot
plt.savefig('./outputs/diabetes_pairplot.png', dpi = 300) 

# Display the plot
plt.show()
```

**Interpretation:**

- Positively correlated, those with outcome=1 tend to have higher values

### Plot 8. Heatmap

```python
# 8. Correlation heatmap
corr_mat = df.corr()
sns.heatmap(corr_mat)
```

**Interpretation**
- Strong positive correlation between BMI & Skin Thickness and Insulin and Skin Thickness

# Data Visulization on 'Women Ecommerce Clothing Reviews' using `matplotlib`

- The data can be downloaded from the following Kaggle [link](https://www.kaggle.com/datasets/nicapotato/womens-ecommerce-clothing-reviews)

## About the Dataset

The dataset seems to be about women's clothing e-commerce reviews. It has several columns:

- Unnamed: 0: Index or identifier.
- Clothing ID: Identifier for the clothing item.
- Age: Age of the reviewer.
- Title: Title of the review.
- Review Text: The actual text of the review.
- Rating: The rating given by the user (on a scale, presumably of 1 to 5).
- Recommended IND: Binary indicator suggesting whether the reviewer recommends the product or not (1 for Yes, 0 for No).
- Positive Feedback Count: Count of positive feedbacks for the review.
- Division Name: Division name for the clothing.
- Department Name: Department name for the clothing.
- Class Name: Class name for the clothing.

- Let's create the visualizations to get deep insight the data using different plots

```python
# Import the libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
ImportWarning('ignore')
```

```python
# Load the dataset
df = pd.read_csv('./data/Womens Clothing E-Commerce Reviews.csv', index_col=0)
```

### Plot 1. Histogram:

```python
# Set style
plt.style.use('seaborn-whitegrid')

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the distribution of ages of reviewers
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')

# Set the x and y-axis labels
plt.xlabel('Age')
plt.ylabel('Number of Reviews')

# Set the plot title
plt.title('Distribution of Ages of Reviewers')

# Save the plot 
plt.savefig('./outputs/plots/hist_plot_EC.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

**Plot 1 Interpretation:**

The distribution of ages of reviewers shows a right-skewed pattern. Most reviewers are in the age range of 30 to 50 years, with the peak being around the mid-30s.

### Plot 2. Barplot

-  let's visualize the distribution of ratings given by the reviewers.​

```python
# Plot the distribution of ratings

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the distribution of ratings
df['Rating'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')

# Set the x and y-axis labels
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')

# Set the plot title
plt.title('Distribution of Ratings')

# Save the plot
plt.savefig('./outputs/plots/bar_plot_EC.png', dpi=300, bbox_inches='tight')

# Set the x-axis labels to be horizontal
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

**Plot 2 Interpretation:**

The majority of the reviews have a rating of 5, followed by 4. Ratings of 1 and 2 are the least frequent, indicating that most customers seem satisfied with their purchases.

### Plot 3. Heatmap
- Let's start with the heatmap to show correlations between numeric variables.

```python
# Compute the correlation matrix
corr = df[['Age', 'Rating', 'Recommended IND', 'Positive Feedback Count']].corr()

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)

# Set the plot title
plt.title('Heatmap of Correlation between Numeric Variables')

# Save the plot
plt.savefig('./outputs/plots/heatmap_EC.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

**Heatmap Interpretation:**
- The heatmap shows the correlation between numeric variables. There's a slight positive correlation between "Rating" and "Recommended IND", which makes intuitive sense: higher ratings are more likely to be associated with product recommendations. Other correlations are relatively weak.

### Plot 4. Line Chart
- let's create a line chart to display the average rating by age

```python
# Calculate average rating by age
avg_rating_by_age = df.groupby('Age')['Rating'].mean()

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the line chart
avg_rating_by_age.plot(kind='line', color='skyblue', marker='o')

# Set the x and y-axis labels
plt.xlabel('Age')
plt.ylabel('Average Rating')

# Set the plot title
plt.title('Line Chart of Average Rating by Age')

# Save the plot
plt.savefig('./outputs/plots/line_plot_EC.png', dpi=300, bbox_inches='tight')

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.5) 

# Display the plot
plt.show()
```

**Line Chart Interpretation:**
- The line chart displays the average rating given by reviewers of different ages. 
- While the ratings fluctuate across ages, they generally remain above 4. 
- Some specific ages show spikes or dips, but these could be due to fewer reviews from those age groups, leading to more variability.

### Plot 5. Area Chart
- let's create an area chart to display the cumulative positive feedback count by age.​

```python
# Calculate cumulative positive feedback count by age
cumulative_feedback_by_age = df.groupby('Age')['Positive Feedback Count'].sum().cumsum()

# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the area chart with green color
cumulative_feedback_by_age.plot(kind='area', color='green', alpha=0.5)

# Set the x and y-axis labels
plt.xlabel('Age')
plt.ylabel('Cumulative Positive Feedback Count')

# Set the plot title
plt.title('Area Chart of Cumulative Positive Feedback by Age')

# Save the plot
plt.savefig('./outputs/plots/area_plot_EC.png', dpi=300, bbox_inches='tight')

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.5) 

# Display the plot
plt.show()
```

**Area Chart Interpretation:**

- The area chart visualizes the cumulative positive feedback count by age. As age increases, the cumulative count of positive feedbacks also rises. 
- This indicates that older reviewers, in aggregate, have received more positive feedbacks over time, which could be due to a higher number of reviews or more resonating reviews from these age groups.

### Plot 6. Pie Chart
- Let's create a pie chart to visualize the distribution of recommendations.


```python
# Prepare data for the pie chart
recommendation_counts = df['Recommended IND'].value_counts()
labels = ['Recommended', 'Not Recommended']
sizes = recommendation_counts.values
colors = ['skyblue', 'salmon']
explode = (0.1, 0)  # explode 1st slice for emphasis

# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Set the plot title
plt.title('Pie Chart of Recommendations Distribution')

# Save the plot
plt.savefig('./outputs/plots/pie_plot_EC.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

**Pie Chart Interpretation:**

A significant majority (around 82%) of the reviews recommend the products, while only about 18% do not recommend them.

### plot 7. Barplot (distribution of recommendations)
- let's visualize the distribution of recommendations, distinguishing between products that were recommended and those that were not.​

```python
# Plot the distribution of recommendations

# Set the figure size
plt.figure(figsize=(8, 6))

# Plot the distribution of recommendations using barplot
df['Recommended IND'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'], edgecolor='black')

# Set the x and y-axis labels
plt.xlabel('Recommendation (1: Recommended, 0: Not Recommended)')
plt.ylabel('Number of Reviews')

# Set the plot title
plt.title('Distribution of Recommendations')

# Save the plot
plt.savefig('./outputs/plots/bar_plot2_EC.png', dpi=300, bbox_inches='tight')

# Set the x-axis labels to be horizontal
plt.xticks(rotation=0) 

# Save the plot
plt.show()
```

**Barplot (distribution of Recommendations) Interpretation:** 
- A significant majority of the reviews recommend the products, as indicated by the larger bar for "1" (Recommended). This suggests that the overall sentiment towards the products in this dataset is positive.

### Plot 8. Barplot for distribution of positive feedback count 
- let's visualize the distribution of the positive feedback count received by the reviews.

```python
# Plot the distribution of positive feedback count

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the distribution of positive feedback count
plt.hist(df['Positive Feedback Count'], bins=50, color='green', edgecolor='black', range=[0, 20])

# Set the x and y-axis labels
plt.xlabel('Positive Feedback Count')
plt.ylabel('Number of Reviews')

# Set the plot title
plt.title('Distribution of Positive Feedback Count (Capped at 20)')

# Save the plot
plt.savefig('./outputs/plots/hist_plot3_EC.png', dpi=300, bbox_inches='tight')

plt.show()
```

**Interpretation:**
- Most reviews have received very few positive feedbacks, with a large portion receiving none at all.
- Only a small fraction of reviews have garnered more than 5 positive feedbacks. This distribution is highly skewed to the right.

### Plot 9. Barplot (Distribution of reviews across various Departments​)
- Let's now visualize the distribution of reviews across various departments.​

```python
# Plot the distribution of reviews across departments

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the distribution of reviews across departments
df['Department Name'].value_counts().plot(kind='bar', color='purple', edgecolor='black')

# Set the x and y-axis labels
plt.xlabel('Department Name')
plt.ylabel('Number of Reviews')

# Set the plot title
plt.title('Distribution of Reviews across Departments')

# Save the plot
plt.savefig('./outputs/plots/bar_plot4_EC.png', dpi=300, bbox_inches='tight')

# Set the x-axis labels to be inclined at 45 degrees
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

### Plot 10. Bar Plot(Average rating by Department​)
- let's now visualize the average rating by department.​

```python
# Calculate average rating by department
avg_rating_by_department = df.groupby('Department Name')['Rating'].mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Plot the average rating by department
avg_rating_by_department.plot(kind='bar', color='pink', edgecolor='black')

# Set the x and y-axis labels
plt.xlabel('Department Name')
plt.ylabel('Average Rating')

# Set the plot title
plt.title('Average Rating by Department')

# Save the plot
plt.savefig('./outputs/plots/bar_plot5_EC.png', dpi=300, bbox_inches='tight')

plt.ylim(3, 5)  # Set y-axis limits to better visualize differences
plt.xticks(rotation=45) # Set the x-axis labels to be inclined at 45 degrees

# Display the plot
plt.show()
```

**Interpretation:**
- All departments have an average rating above 4, with subtle differences between them.
